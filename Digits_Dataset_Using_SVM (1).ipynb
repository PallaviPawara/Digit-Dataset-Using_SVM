#### Split data into Train and Test :

```python
from sklearn.model_selection import train_test_split
```

```python
df_train , df_test = train_test_split(df , test_size = .2 ,random_state = 555)
```

```python
df_train_x = df_train.iloc[: , 0:-1]
df_train_y = df_train.iloc[: , -1]
```

```python
df_test_x = df_test.iloc[: , 0:-1]
df_test_y = df_test.iloc[: , -1]
```

```python
print(df_train_x.shape)
print(df_train_y.shape)
```

```python
print(df_test_x.shape)
print(df_test_y.shape)
```

#### Import SVM from sklearn Library :

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(df_train_x , df_train_y)
```

```python
model.score(df_train_x , df_train_y)
```

##### Prediction on test data :


```python
pred_value = model.predict(df_test_x)
```

```python
pred_value
```

#### Confusion Matrix :

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(pred_value , df_test_y)
```

```python
cm
```

#### Efficiency:

```python
cm.diagonal().sum()*100/cm.sum()
```

### Tune Parameters :

#### 1] Regularization(C) :

```python
model_c = SVC(C = 1)
```

```python
model_c.fit(df_train_x , df_train_y)
```

```python
model_c.score(df_test_x , df_test_y)
```

```python
model_c = SVC(C = 10)
```

```python
model_c.fit(df_train_x , df_train_y)
```

```python
model_c.score(df_test_x , df_test_y)
```

#### 2] Gamma :


```python
model_g = SVC(gamma = 10)
```

```python
model_g.fit(df_train_x , df_train_y)
```

```python
model_g.score(df_test_x , df_test_y)
```

 #### 3] Kernel :

```python
model_linear_kernel = SVC(kernel = "linear")
```

```python
model_linear_kernel.fit(df_train_x , df_train_y)
```

```python
model_linear_kernel.score(df_test_x , df_test_y)
```

```python
model_linear_kernel_rbf = SVC(kernel = "rbf")
```

```python
model_linear_kernel_rbf.fit(df_train_x , df_train_y)
```

```python
model_linear_kernel_rbf.score(df_test_x , df_test_y)
```
